```python
#Read table as pandas dataframe from wiki
import pandas as pd
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]

#remove header row
df.drop(df.index[0], inplace=True)

#rename column
df.rename(columns={0: "PostalCode", 1: "Borough", 2:"Neighborhood"}, inplace=True)
```

```python
#Filter out 'Not assigned' borough
df_filtered = df[df['Borough']!='Not assigned'].copy()
#remove dummy line return character from web page data
#type(df_filtered)
df_filtered["Neighborhood"] = df_filtered["Neighborhood"].replace(r'\n','', regex=True) 
```

```python
#Don't see such data from wiki, but include it for completeness
#Replace the neighborhood value with borough if its value is 'Not assigned'
import numpy as np
df_filtered['Neighborhood'] = np.where(df_filtered['Neighborhood']=='Not assigned', df_filtered['Borough'], df_filtered['Neighborhood'])
```

```python
#concatenate Neighborhood group by code and borough
df_result = pd.DataFrame(df_filtered.groupby(['PostalCode', 'Borough'])['Neighborhood'].agg(lambda col: ', '.join(col)))
df_result.reset_index(inplace=True)
df_result
```

```python
#explore top 100 venues next to each neighborhood
radius=500
LIMIT=100

#borrow functions from lab

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#function that extracts venues from different locations
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```

```python
toronto_venues.groupby('Neighborhood').count()
```

```python
print('There are {} uniques categories.'.format(len(toronto_venues['Venue Category'].unique())))
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
#use 'Neighborhood_name' instead as there is a Venue Category named 'Neighborhood'
toronto_onehot['Neighborhood_name'] = toronto_venues['Neighborhood'] 
# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head(20)
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood_name').mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

```python
# set number of clusters
kclusters = 5

toronto_neighborhoods = pd.DataFrame()
toronto_neighborhoods['Neighborhood'] = toronto_grouped['Neighborhood_name']

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood_name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
toronto_neighborhoods.insert(0, 'Cluster Labels', kmeans.labels_)

#toronto_grouped
# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_data.join(toronto_neighborhoods.set_index('Neighborhood'), on='Neighborhood')
#toronto_neighborhoods
toronto_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters

#it can be observed that most neighborhoods belong to the same cluster
#Neighborhoods with different cluster are normally far from the city center.
```
